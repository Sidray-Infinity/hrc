- Multiple columns in the dataset are completely NULL. Need to remove them.

```python
# droping completely null columns

count = 0
for col in df.columns:
    if df[col].isnull().sum() == df.shape[0]:
        count += 1
        df.drop(col, axis=1, inplace=True)
        
print("Num columns droped:", count)

print(df.columns, df.shape)
```

- Duplicate columns/features exists in the dataset. Removing them, as they are just redundant information 

```python
def duplicate_pairs(df):
    """
    Return a set of duplicate pairs
    """
    
    duplicate_pairs = []

    for i in range(df.shape[1]):
        for j in range(i+1, df.shape[1]):
            if (df.iloc[:, i] == df.iloc[:, j]).sum() == df.shape[0]:
                duplicate_pairs.append([df.columns[i], df.columns[j]])


    for i in range(len(duplicate_pairs)):
        for ele in duplicate_pairs[i]:
            for j in range(i+1, len(duplicate_pairs)):
                if ele in duplicate_pairs[j]:
                    duplicate_pairs[j].remove(ele)
                    duplicate_pairs[i].extend(duplicate_pairs[j])

    
    duplicate_pairs = [tuple(set(x)) for x in duplicate_pairs if len(x) > 1]
    
    return duplicate_pairs
```

- Find the duplicate pairs.
- Go through the list of duplicate pairs, and select the features to be dropped
- Not automating this process, because domain knowledge is required while selecting a feature to drop
- Enter -1 if you do not want to drop any of the column shown

```python
"""
[IMPORTANT]: RUN THIS CELL TWICE TO REMOVE ALL DUPLICATE COLUMNS
- When asked for input, enter the index of the element in the tuple, representing the column name to remove
- To be removed : 
    * acct_doc_header_id
    * invoice_id, doc_number
    * cust_number
    * document_create_date
    * total_open_amount
    * invoice_currency
    * doc_number
     
"""

pairs = duplicate_pairs(df)

for x in pairs:
    print(x)
    indices = list(map(int, input("Enter the indices of features to drop(0 based): ").split(' ')))
    for i in indices:
        if i == -1:
            break
        df.drop(x[i], axis=1, inplace=True)

print(df.columns)
df.head()
```

The four cells below encode the day, month, year and payment terms column based on the mean of their corresponding actual_outstanding payment values.

```python
# Encoding doc create month, daym year and payment terms as well

mapper_year = train_x.groupby('doc_create_year')['actual_outstanding_amount'].mean().to_dict()
train_x['doc_create_year_encoded'] = train_x.doc_create_year.map(mapper_year)
test_x['doc_create_year_encoded'] = test_x.doc_create_year.map(mapper_year)
```

```python
mapper_month = train_x.groupby('doc_create_month')['actual_outstanding_amount'].mean().to_dict()
train_x['doc_create_month_encoded'] = train_x.doc_create_month.map(mapper_month)
test_x['doc_create_month_encoded'] = test_x.doc_create_month.map(mapper_month)
```

```python
mapper_day = train_x.groupby('doc_create_day')['actual_outstanding_amount'].mean().to_dict()
train_x['doc_create_day_encoded'] = train_x.doc_create_day.map(mapper_day)
test_x['doc_create_day_encoded'] = test_x.doc_create_day.map(mapper_day)
```

```python
mapper_pt = train_x.groupby('cust_payment_terms')['actual_outstanding_amount'].mean().to_dict()
train_x['cust_payment_terms_encoded'] = train_x.cust_payment_terms.map(mapper_pt)
test_x['cust_payment_terms_encoded'] = test_x.cust_payment_terms.map(mapper_pt)
```
